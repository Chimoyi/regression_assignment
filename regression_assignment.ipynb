We can also check for outliers, for instance the max value in MEDV is 50 which is far off from the min 22.532806, so we can check for the percentage of outliers and remove the ones above 50

```python
#following the report on Kaggle

for k, v in data.items():
        q1 = v.quantile(0.25)
        q3 = v.quantile(0.75)
        irq = q3 - q1
        v_col = v[(v <= q1 - 1.5 * irq) | (v >= q3 + 1.5 * irq)]
        perc = np.shape(v_col)[0] * 100.0 / np.shape(data)[0]
        print("Column %s outliers = %.2f%%" % (k, perc))
        
#deleting the outliers in MEDV
data = data[~(data['MEDV'] >= 50.0)]
np.shape(data)
```
